```python
import numpy as np
import matplotlib
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation, rc
from IPython.display import HTML
import pgmpy as pgm
from sklearn.decomposition import PCA
import useful
from sklearn.manifold import LocallyLinearEmbedding

%matplotlib notebook
#matplotlib.rcParams['figure.figsize'] = [16, 9]
#matplotlib.rcParams['figure.dpi'] = 150
```

```python
data = np.load('../chorales_vectors_12.npy')
data.shape
```

```python
pca = PCA(n_components=100)
data_tr = pca.fit_transform(data)
```

```python
data_tr.shape
```

```python
import useful

useful.scatter_3d(
    data_tr[:, :3].T, 
    rcParams=matplotlib.rcParams)
```

Here, we will perform a different dimensionality reduction technique because the data exists in a non-linear manifold.

```python
# Create a LocallyLinearEmbedding Object

LLE = LocallyLinearEmbedding(n_components = 3)
data_LLE = LLE.fit_transform(data)
data_LLE[:5,:]
```
